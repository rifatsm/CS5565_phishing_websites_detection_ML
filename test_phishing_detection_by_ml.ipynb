```python
import numpy as np  
import matplotlib.pyplot as plt
import pandas as pd

# Read the data
data = pd.read_csv('data/dataset.csv')

# Discarding the 'index' column
del data['index']

# View first and last 5 observations
print(data.head())
print(data.tail())
```

```python
# Describe statistical information of data
print(data.describe())
# Below stats show that 75 percentile of obseravtions belong to class 1 
```

```python
# Check column types
print(data.info())               

# All comumns are int type, so no change is required
```

```python
# Plot distribution of classes using Histograms
plt.figure(figsize =(8,8))
plt.hist(data.Result)           

# It shows that benign class have about 1000+ observations than malware
```

```python
# Look for missing values
print(data.isnull().sum())        

# No missing values found, so no need to drop or replace any value
```

```python
# Generate correlation matrix
print(data.corr())

import seaborn as sns
plt.figure(figsize =(8,8))
sns.heatmap(data.corr())    # Generate heatmap (though very less clarity due to large no. of ftrs
```

```python
print(data.corr()['Result'].sort_values())      # Print correlation with target variable
```

```python
# Remove features having correlation coeff. between +/- 0.03
data.drop(['Favicon','Iframe','Redirect',
                'popUpWidnow','RightClick','Submitting_to_email'],axis=1,inplace=True)
print(len(data.columns))
```

```python
# Prepare data for models
y = data['Result'].values
X = data.drop(['Result'], axis = 1)

from sklearn.metrics import accuracy_score,roc_curve,auc, confusion_matrix, precision_recall_fscore_support
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

# Split the data as training and testing data - 70% train size, 30% test size
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size = 0.3, random_state = None)
```

```python
from datetime import datetime

# dt_a = datetime.now().microsecond

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
%time rfc = rfc.fit(X_train,y_train)
%time prediction = rfc.predict(X_test)
print("Accuracy with Random Forest classifier:",accuracy_score(y_test, prediction)) 
fpr,tpr,thresh = roc_curve(y_test,prediction)      
roc_auc = accuracy_score(y_test,prediction)         # Calculate ROC AUC

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for Random Forest
plt.plot(fpr,tpr,'green',label = 'Random Forest')
plt.legend("Random Forest", loc='lower right')
plt.legend(loc='lower right')
print("Conf matrix Random Forest classifier:")  #  Generate confusion matrix
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))

#2 Classification using logistic regression
# dt_a = datetime.now().microsecond

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
%time logreg = logreg.fit(X_train,y_train)
%time prediction = logreg.predict(X_test)
print("Accuracy with Logistic Regression:", accuracy_score(y_test, prediction))
print ("Conf matrix Logistic Regression:")
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))


fpr,tpr,thresh = roc_curve(y_test,prediction)
roc_auc = accuracy_score(y_test,prediction)

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for Logistic Regression
plt.plot(fpr,tpr,'orange',label = 'Logistic Regression')
plt.legend("Logistic Regression", loc='lower right')
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.legend(loc='lower right')

#3 Classification using SVM
# dt_a = datetime.now().microsecond

from sklearn.svm import SVC
svc_l = SVC(kernel = "linear", C = 0.025)
%time svc_l = svc_l.fit(X_train,y_train)
%time prediction = svc_l.predict(X_test)
print("Accuracy with SVM-Linear:",accuracy_score(y_test, prediction))
fpr,tpr,thresh = roc_curve(y_test,prediction)
roc_auc = accuracy_score(y_test,prediction)

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for SVM-linear
plt.plot(fpr,tpr,'blue',label = 'SVM')
plt.legend("SVM", loc ='lower right')
plt.legend(loc ='lower right')
print("Conf matrix SVM-linear:")
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))


#4 Classification using NearestCentroid
# Link: https://scikit-learn.org/stable/modules/neighbors.html#nearest-neighbors-classification
# dt_a = datetime.now().microsecond

from sklearn.neighbors import NearestCentroid
nc_l = NearestCentroid()
%time nc_l = nc_l.fit(X_train,y_train)
%time prediction = svc_l.predict(X_test)
print("Accuracy with NearestCentroid:",accuracy_score(y_test, prediction))
fpr,tpr,thresh = roc_curve(y_test,prediction)
roc_auc = accuracy_score(y_test,prediction)

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for NearestCentroid
plt.plot(fpr,tpr,'red',label = 'NearestCentroid')
plt.legend("NearestCentroid", loc ='lower right')
plt.legend(loc ='lower right')
print("Conf matrix NearestCentroid:")
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))

#6 Classification using MultiLayerPerceptron
# dt_a = datetime.now().microsecond

# Link: https://scikit-learn.org/stable/modules/neural_networks_supervised.html#multi-layer-perceptron
from sklearn.neural_network import MLPClassifier
mlc = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(5, 2), random_state=1)
%time mlc = mlc.fit(X_train,y_train)
%time prediction = mlc.predict(X_test)
print("Accuracy with MultiLayerPerceptron:",accuracy_score(y_test, prediction))

fpr,tpr,thresh = roc_curve(y_test,prediction)
roc_auc = accuracy_score(y_test,prediction)

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for SVM-linear
plt.plot(fpr,tpr,'purple',label = 'MultiLayerPerceptron')
plt.legend("MultiLayerPerceptron", loc ='lower right')
plt.legend(loc ='lower right')
print("Conf matrix MultiLayerPerceptron:")
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))

#6.2 Classification using MultiLayerPerceptron
# dt_a = datetime.now().microsecond

# Link: https://scikit-learn.org/stable/modules/neural_networks_supervised.html#multi-layer-perceptron
from sklearn.neural_network import MLPClassifier
mlc2 = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(15,), random_state=1)
%time mlc2 = mlc2.fit(X_train,y_train)
%time prediction = mlc2.predict(X_test)
print("Accuracy with MultiLayerPerceptron2:",accuracy_score(y_test, prediction))


fpr,tpr,thresh = roc_curve(y_test,prediction)
roc_auc = accuracy_score(y_test,prediction)

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for SVM-linear
plt.plot(fpr,tpr,'pink',label = 'MultiLayerPerceptron2')
plt.legend("MultiLayerPerceptron2", loc ='lower right')
plt.legend(loc ='lower right')
print("Conf matrix MultiLayerPerceptron2:")
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))

# 7 Classification using DecisionTreeClassifier
# dt_a = datetime.now().microsecond

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
%time dtc = dtc.fit(X_train,y_train)
%time prediction = dtc.predict(X_test)
print("Accuracy with DecisionTreeClassifier:",accuracy_score(y_test, prediction)) 


fpr,tpr,thresh = roc_curve(y_test,prediction)      
roc_auc = accuracy_score(y_test,prediction)         # Calculate ROC AUC

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for DecisionTreeClassifier
plt.plot(fpr,tpr,'gray',label = 'DecisionTreeClassifier')
plt.legend("DecisionTreeClassifier", loc='lower right')
plt.legend(loc='lower right')
print("Conf matrix DecisionTreeClassifier:")  #  Generate confusion matrix
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))


# 8 Classification using ExtraTreesClassifier
# dt_a = datetime.now().microsecond

# Best (so far)
from sklearn.ensemble import ExtraTreesClassifier
clf_xtree = ExtraTreesClassifier()
%time clf_xtree = clf_xtree.fit(X_train,y_train)
%time prediction = clf_xtree.predict(X_test)
print("Accuracy with ExtraTreesClassifier:",accuracy_score(y_test, prediction)) 
fpr,tpr,thresh = roc_curve(y_test,prediction)      
roc_auc = accuracy_score(y_test,prediction)         # Calculate ROC AUC

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for ExtraTreesClassifier
plt.plot(fpr,tpr,'black',label = 'ExtraTreesClassifier')
plt.legend("ExtraTreesClassifier", loc='lower right')
plt.legend(loc='lower right')
print("Conf matrix ExtraTreesClassifier:")  #  Generate confusion matrix
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))

# 9 Classification using AdaBoostClassifier
# dt_a = datetime.now().microsecond

from sklearn.ensemble import AdaBoostClassifier
clf_ada = AdaBoostClassifier()
%time clf_ada = clf_ada.fit(X_train,y_train)
%time prediction = clf_ada.predict(X_test)
print("Accuracy with AdaBoostClassifier:",accuracy_score(y_test, prediction)) 
fpr,tpr,thresh = roc_curve(y_test,prediction)      
roc_auc = accuracy_score(y_test,prediction)         # Calculate ROC AUC

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for AdaBoostClassifier
plt.plot(fpr,tpr,'teal',label = 'AdaBoostClassifier')
plt.legend("AdaBoostClassifier", loc='lower right')
plt.legend(loc='lower right')
print("Conf matrix AdaBoostClassifier:",confusion_matrix(y_test,prediction))  #  Generate confusion matrix
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))


# 10 Classification using GradientBoostingClassifier
# dt_a = datetime.now().microsecond

from sklearn.ensemble import GradientBoostingClassifier
clf_gb = GradientBoostingClassifier()
%time clf_gb = clf_gb.fit(X_train,y_train)
%time prediction = clf_gb.predict(X_test)
print("Accuracy with GradientBoostingClassifier:",accuracy_score(y_test, prediction)) 
fpr,tpr,thresh = roc_curve(y_test,prediction)      
roc_auc = accuracy_score(y_test,prediction)         # Calculate ROC AUC

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for GradientBoostingClassifier
plt.plot(fpr,tpr,'black',label = 'GradientBoostingClassifier')
plt.legend("GradientBoostingClassifier", loc='lower right')
plt.legend(loc='lower right')
print("Conf matrix GradientBoostingClassifier:")  #  Generate confusion matrix
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))

# 11 Classification using HistGradientBoostingClassifier
# dt_a = datetime.now().microsecond

from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier
clf_hist = HistGradientBoostingClassifier()
%time clf_hist = clf_hist.fit(X_train,y_train)
%time prediction = clf_hist.predict(X_test)
print("Accuracy with HistGradientBoostingClassifier:",accuracy_score(y_test, prediction)) 
fpr,tpr,thresh = roc_curve(y_test,prediction)      
roc_auc = accuracy_score(y_test,prediction)         # Calculate ROC AUC

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for HistGradientBoostingClassifier
plt.plot(fpr,tpr,'brown',label = 'HistGradientBoostingClassifier')
plt.legend("HistGradientBoostingClassifier", loc='lower right')
plt.legend(loc='lower right')
print("Conf matrix HistGradientBoostingClassifier:")  #  Generate confusion matrix
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))

# 12 Classification using GaussianNB
# dt_a = datetime.now().microsecond

from sklearn.naive_bayes import GaussianNB
clf_gnb = GaussianNB()
%time clf_gnb = clf_gnb.fit(X_train,y_train)
%time prediction = clf_gnb.predict(X_test)
print("Accuracy with GaussianNB:",accuracy_score(y_test, prediction)) 
fpr,tpr,thresh = roc_curve(y_test,prediction)      
roc_auc = accuracy_score(y_test,prediction)         # Calculate ROC AUC

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for GaussianNB
plt.plot(fpr,tpr,'yellow',label = 'GaussianNB')
plt.legend("GaussianNB", loc='lower right')
plt.legend(loc='lower right')
print("Conf matrix GaussianNB:")  #  Generate confusion matrix
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))

# 13 Classification using VotingClassifier
# dt_a = datetime.now().microsecond

from sklearn.ensemble import VotingClassifier
clf_vote = VotingClassifier(estimators=[('lr', logreg), ('rf', rfc), ('gnb', clf_gnb)],voting='hard')
%time clf_vote = clf_vote.fit(X_train,y_train)
%time prediction = clf_vote.predict(X_test)
print("Accuracy with VotingClassifier:",accuracy_score(y_test, prediction)) 
fpr,tpr,thresh = roc_curve(y_test,prediction)      
roc_auc = accuracy_score(y_test,prediction)         # Calculate ROC AUC

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for VotingClassifier
plt.plot(fpr,tpr,'khaki',label = 'VotingClassifier')
plt.legend("VotingClassifier", loc='lower right')
plt.legend(loc='lower right')
print("Conf matrix VotingClassifier:")  #  Generate confusion matrix
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))

# 14 Classification using GridSearchCV
# dt_a = datetime.now().microsecond

from sklearn.model_selection import GridSearchCV
params = {'lr__C': [1.0, 100.0], 'rf__n_estimators': [20, 200]}
clf_grid = GridSearchCV(estimator=clf_vote, param_grid=params, cv=5)
%time clf_grid = clf_grid.fit(X_train,y_train)
%time prediction = clf_grid.predict(X_test)
print("Accuracy with GridSearchCV:",accuracy_score(y_test, prediction)) 
fpr,tpr,thresh = roc_curve(y_test,prediction)      
roc_auc = accuracy_score(y_test,prediction)         # Calculate ROC AUC

# dt = datetime.now().microsecond - dt_a

# print('elapsed time: '+str(dt))

# Plot ROC curve for GridSearchCV
plt.plot(fpr,tpr,'black',label = 'GridSearchCV')
plt.legend("GridSearchCV", loc='lower right')
plt.legend(loc='lower right')
print("Conf matrix GridSearchCV:")  #  Generate confusion matrix
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))

# # 15 Classification using GaussianMixture [doesn't work]
# from sklearn.mixture import GaussianMixture
# clf_rcv = GaussianMixture(init_params='kmeans')
# clf_rcv = clf_rcv.fit(X_train,y_train)
# prediction = clf_rcv.predict(X_test)
# print("Accuracy with GaussianMixture:",accuracy_score(y_test, prediction)) 
# fpr,tpr,thresh = roc_curve(y_test,prediction)      
# roc_auc = accuracy_score(y_test,prediction)         # Calculate ROC AUC

# # Plot ROC curve for GaussianMixture
# plt.plot(fpr,tpr,'black',label = 'GaussianMixture')
# plt.legend("GaussianMixture", loc='lower right')
# plt.legend(loc='lower right')
# print("Conf matrix GaussianMixture:",confusion_matrix(y_test,prediction))  #  Generate confusion matrix

# # 5 Classification using NeighborhoodComponentsAnalysis
# # Takes way too much time
# # 2nd best
dt_a = datetime.now().microsecond


from sklearn.neighbors import (NeighborhoodComponentsAnalysis, KNeighborsClassifier)
from sklearn.pipeline import Pipeline
nca = NeighborhoodComponentsAnalysis() # random_state=42
knn = KNeighborsClassifier(n_neighbors=10)
nca_pipe = Pipeline([('nca', nca), ('knn', knn)])
%time nca_pipe = nca_pipe.fit(X_train, y_train)
%time prediction = nca_pipe.predict(X_test)
print("Accuracy with NeighborhoodComponentsAnalysis:",accuracy_score(y_test, prediction))
fpr,tpr,thresh = roc_curve(y_test,prediction)
roc_auc = accuracy_score(y_test,prediction)

dt = datetime.now().microsecond - dt_a

print('elapsed time: '+str(dt))

# Plot ROC curve for SVM-linear
plt.plot(fpr,tpr,'magenta',label = 'NeighborhoodComponentsAnalysis')
plt.legend("NeighborhoodComponentsAnalysis", loc ='lower right')
plt.legend(loc ='lower right')
print("Conf matrix NeighborhoodComponentsAnalysis:")
print(confusion_matrix(y_test,prediction))
print("Precision Recall F-score: ")
print(precision_recall_fscore_support(y_test,prediction,average=None))


# Show all the plots together
plt.show()


'''
# -------- Apply Recursive Feature Elimination(RFE) and use reduced feature set for prediction ------------------------
# Recursive Feature Elimination(RFE) is a technique that takes entire feature set as input and removes features one at 
# a time up to a specified number or until a stopping criteria is met.
'''
from sklearn.feature_selection import RFE
rfe = RFE(rfc,27)                              
rfe = rfe.fit(X_train, y_train)               # Train RF classifier with only 27 features now
pred = rfe.predict(X_test)

# Test accuracy on reduced data
print("Accuracy by RFClassifier after RFE is applied:", accuracy_score(y_test,pred))

rfe = RFE(svc_l,27)
rfe = rfe.fit(X_train, y_train)               # Train SVM with only 27 features now
pred = rfe.predict(X_test)
print("Accuracy by SVM-Linear after RFE is applied:", accuracy_score(y_test,pred))

rfe = RFE(logreg,27)
rfe = rfe.fit(X_train, y_train)              # Train Logistic-Reg with only 27 features now
pred = rfe.predict(X_test)
print("Accuracy by Logistic Regression after RFE is applied:", accuracy_score(y_test,pred))


rfe = RFE(nc_l,27)
rfe = rfe.fit(X_train, y_train)              # Train NearestCentroid with only 27 features now
pred = rfe.predict(X_test)
print("Accuracy by NearestCentroid after RFE is applied:", accuracy_score(y_test,pred))


rfe = RFE(mlc,27)                              
rfe = rfe.fit(X_train, y_train)               # Train RF classifier with only 27 features now
pred = rfe.predict(X_test)


# Test accuracy on reduced data
print("Accuracy by MultiLayerPerceptron after RFE is applied:", accuracy_score(y_test,pred))

rfe = RFE(mlc2,27)                              
rfe = rfe.fit(X_train, y_train)               # Train RF classifier with only 27 features now
pred = rfe.predict(X_test)

# Test accuracy on reduced data
print("Accuracy by MultiLayerPerceptron2 after RFE is applied:", accuracy_score(y_test,pred))

rfe = RFE(dtc,27)                              
rfe = rfe.fit(X_train, y_train)               # Train RF classifier with only 27 features now
pred = rfe.predict(X_test)

# Test accuracy on reduced data
print("Accuracy by DecisionTreeClassifier after RFE is applied:", accuracy_score(y_test,pred))

rfe = RFE(clf_xtree,27)                              
rfe = rfe.fit(X_train, y_train)               # Train RF classifier with only 27 features now
pred = rfe.predict(X_test)

# Test accuracy on reduced data
print("Accuracy by ExtraTreesClassifier after RFE is applied:", accuracy_score(y_test,pred))

rfe = RFE(clf_ada,27)                              
rfe = rfe.fit(X_train, y_train)               # Train RF classifier with only 27 features now
pred = rfe.predict(X_test)

# Test accuracy on reduced data
print("Accuracy by AdaBoostClassifier after RFE is applied:", accuracy_score(y_test,pred))

rfe = RFE(clf_gb,27)                              
rfe = rfe.fit(X_train, y_train)               # Train RF classifier with only 27 features now
pred = rfe.predict(X_test)

# Test accuracy on reduced data
print("Accuracy by GradientBoostingClassifier after RFE is applied:", accuracy_score(y_test,pred))

rfe = RFE(clf_hist,27)                              
rfe = rfe.fit(X_train, y_train)               # Train RF classifier with only 27 features now
pred = rfe.predict(X_test)

# Test accuracy on reduced data
print("Accuracy by HistGradientBoostingClassifier after RFE is applied:", accuracy_score(y_test,pred))

rfe = RFE(clf_gnb,27)                              
rfe = rfe.fit(X_train, y_train)               # Train RF classifier with only 27 features now
pred = rfe.predict(X_test)

# Test accuracy on reduced data
print("Accuracy by GaussianNB after RFE is applied:", accuracy_score(y_test,pred))

rfe = RFE(clf_vote,27)                              
rfe = rfe.fit(X_train, y_train)               # Train RF classifier with only 27 features now
pred = rfe.predict(X_test)

# Test accuracy on reduced data
print("Accuracy by VotingClassifier after RFE is applied:", accuracy_score(y_test,pred))

rfe = RFE(clf_grid,27)                              
rfe = rfe.fit(X_train, y_train)               # Train RF classifier with only 27 features now
pred = rfe.predict(X_test)

# Test accuracy on reduced data
print("Accuracy by GridSearchCV after RFE is applied:", accuracy_score(y_test,pred))

rfe = RFE(nca_pipe,27)                              
rfe = rfe.fit(X_train, y_train)               # Train RF classifier with only 27 features now
pred = rfe.predict(X_test)

# Test accuracy on reduced data
print("Accuracy by NeighborhoodComponentsAnalysis after RFE is applied:", accuracy_score(y_test,pred))
```
